# **Importing required Libraries**

```python
# Import the profanity module for explicit content detection
from better_profanity import profanity

# Import the matplotlib library for data visualization
import matplotlib.pyplot as plt

# Import the numpy library for numerical operations
import numpy as np
```

# **Defining a function to calculate % of profanity**

```python
# Function to calculate the percentage of profanity in the given text
def calculate_profanity_percentage(text):
    # Split the text into individual words
    words = text.split()

    # Count the total number of words
    total_words = len(words)

    # Count the number of profane words
    profane_words = sum(profanity.contains_profanity(word) for word in words)

    # Calculate the percentage of profanity in the text
    profanity_percentage = (profane_words / total_words) * 100 if total_words > 0 else 0

    # Return the calculated profanity percentage
    return profanity_percentage
```

# **Data Visualisation**

```python
# Function to visualize the distribution of profanity in the text
def visualize_profanity_distribution(text):
    # Split the text into individual words
    words = text.split()

    # Create a list indicating whether each word contains profanity
    profanity_distribution = [profanity.contains_profanity(word) for word in words]

    # Create a bar chart to visualize profanity distribution
    plt.figure(figsize=(10, 3))
    plt.bar(range(len(words)), profanity_distribution, color='red', alpha=0.7)
    plt.title('Profanity Distribution in Text')
    plt.xlabel('Word Index')
    plt.ylabel('Contains Profanity')
    plt.show()
```

# **Censoring Explicit Content & Working of Model**

```python
# Decorate and prompt the user to enter a text for profanity check
print("\n=== Content Filtering System for Parental Control ===")
text = input("\nEnter the text to check for explicit words: ")

# Check if the entered text contains profanity
if profanity.contains_profanity(text):
    print("\nThe text contains profane words.")

    # Censor the profane words in the text
    censored_text = profanity.censor(text)
    print("\nCensored Text:")
    print(censored_text)

    # Calculate and print the profanity percentage in the text
    profanity_percentage = calculate_profanity_percentage(text)
    print(f"\nProfanity Percentage: {profanity_percentage:.2f}%")

    # Highlight profane words in the text for display
    highlighted_text = ' '.join(f'<span style="color:red">{word}</span>' if profanity.contains_profanity(word) else word for word in text.split())
    print("\nHighlighted Text:")
    print(highlighted_text)

    # Visualize profanity distribution in the text
    visualize_profanity_distribution(text)
else:
    print("\nNo, the text does not contain profane words.")
```
